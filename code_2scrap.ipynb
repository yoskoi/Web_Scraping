```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

# Fonction pour extraire les données d'une page donnée
def extract_data_from_page(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    
    # Trouver le tableau sur la page
    table = soup.find("table", class_="sc-14cb040a-3 dsflYb cmc-table")

    # Extraire les en-têtes de colonnes
    headers = table.find_all("th")
    column_names = [header.text.strip() for header in headers]

    # Initialiser une liste pour stocker les données
    crypto_data = []

    # Extraire les données de chaque ligne du tableau
    rows = table.find_all("tr")
    for row in rows[1:]:  # Commencer depuis l'index 1 pour ignorer les en-têtes de colonnes
        row_data = row.find_all("td")
        row_values = [data.text.strip() for data in row_data]
        crypto_data.append(row_values)

    return crypto_data

# Fonction pour extraire les données de toutes les pages
def extract_data_from_all_pages(start_page=1, end_page=5):
    base_url = "https://coinmarketcap.com/?page="
    all_data = []

    for page in range(start_page, end_page + 1):
        url = base_url + str(page)
        data = extract_data_from_page(url)
        all_data.extend(data)

    return all_data

# Extraire les données de toutes les pages
all_crypto_data = extract_data_from_all_pages(start_page=1, end_page=5)

# Créer un DataFrame pandas à partir des données
df = pd.DataFrame(all_crypto_data, columns=["Name", "Price", "1h %", "24h %", "7d %", "Market Cap", "Volume(24h)", "Circulating Supply"])

# Afficher le DataFrame
print(df)
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

def extract_column_names(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    
    # Trouver le tableau sur la page
    table = soup.find("table", class_="sc-14cb040a-3 dsflYb cmc-table")

    # Extraire les en-têtes de colonnes
    headers = table.find_all("th")
    column_names = [header.text.strip() for header in headers if header.text.strip()]  # Filtrer les valeurs vides
    column_names = [name for name in column_names if name != '#']  # Supprimer les en-têtes non pertinents
    return column_names

# URL de la première page de CoinMarketCap
url_c = "https://coinmarketcap.com/"

# Extraire les noms de colonnes de la première page
column_names = extract_column_names(url_c)
print("Noms de colonnes possibles :", column_names)
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

# Fonction pour extraire les données d'une page donnée
def extract_data_from_page(url, column_names):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    
    # Trouver le tableau sur la page
    table = soup.find("table", class_="sc-14cb040a-3 dsflYb cmc-table")

    # Initialiser une liste pour stocker les données
    crypto_data = []

    # Extraire les données de chaque ligne du tableau
    rows = table.find_all("tr")
    for row in rows[1:]:  # Commencer depuis l'index 1 pour ignorer les en-têtes de colonnes
        row_data = row.find_all("td")
        row_values = [data.text.strip() for data in row_data]
        
        # Vérifier si le nombre de colonnes correspond au nombre de noms de colonnes
        if len(row_values) == len(column_names):
            crypto_data.append(row_values)

    return crypto_data

# Fonction pour extraire les données de toutes les pages
def extract_data_from_all_pages(start_page=1, end_page=5):
    base_url = "https://coinmarketcap.com/?page="
    all_data = []
    
    # Extraire les noms de colonnes possibles
    column_names = extract_column_names(base_url + str(start_page))

    for page in range(start_page, end_page + 1):
        url = base_url + str(page)
        data = extract_data_from_page(url, column_names)
        all_data.extend(data)

    return all_data

# Extraire les données de toutes les pages
all_crypto_data = extract_data_from_all_pages(start_page=1, end_page=5)

# Créer un DataFrame pandas à partir des données
df = pd.DataFrame(all_crypto_data, columns=["Name", "Price", "1h %", "24h %", "7d %", "Market Cap", "Volume(24h)", "Circulating Supply", "Last 7 Days"])

# Afficher le DataFrame
print(df)
```
