```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

URL = "https://fr.finance.yahoo.com/crypto-monnaies/"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
table = soup.find("table")

# Extraire les en-têtes de colonne
headers = []
for header in table.find("tr").find_all("th"):
    headers.append(header.text.strip())

# Extraire les données des cellules du tableau
data = []
for row in table.find_all("tr")[1:]:  # On commence à l'index 1 pour exclure les en-têtes de colonne
    row_data = []
    for cell in row.find_all("td"):
        row_data.append(cell.text.strip())
    data.append(row_data)

# Créer un DataFrame à partir des données extraites
df = pd.DataFrame(data, columns=headers)

# Afficher le DataFrame
#print(df)
df
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

URL = "https://fr.finance.yahoo.com/crypto-monnaies/"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
table = soup.find("table")

# Extraire les données des cellules du tableau
data = []
for row in table.find_all("tr"):
    row_data = []
    for cell in row.find_all("td"):
        # Convertir les valeurs numériques en chaînes de caractères
        row_data.append(str(cell.text.strip()))
    data.append(row_data)

# Créer un DataFrame à partir des données extraites
df = pd.DataFrame(data)

# Afficher le DataFrame
print(df)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import csv
from datetime import datetime

def charger_etudiants():
    students = []
    with open('C:/Users/L390/Documents/aaa/tuto_python_2/crypos.csv', mode='r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            students.append(row)
    return students

def crypto_actu():
    URL = "https://fr.finance.yahoo.com/crypto-monnaies/"
    page = requests.get(URL)

    soup = BeautifulSoup(page.content, "html.parser")
    table = soup.find("table")

    # Extraire les en-têtes de colonne
    headers = []
    for header in table.find("tr").find_all("th"):
        headers.append(header.text.strip())

    # Extraire les données des cellules du tableau
    data = []
    for row in table.find_all("tr")[1:]:  # On commence à l'index 1 pour exclure les en-têtes de colonne
        row_data = []
        for cell in row.find_all("td"):
            row_data.append(cell.text.strip())
        data.append(row_data)

    # Créer un DataFrame à partir des données extraites
    df = pd.DataFrame(data, columns=headers)

    # Enregistrer les données dans un fichier CSV avec la date actuelle comme nom de fichier
    current_date = datetime.now().strftime("%Y-%m-%d")
    filename = f"cryptos_{current_date}.csv"
    df.to_csv(filename, index=False)  # Enregistrer le DataFrame dans un fichier CSV sans l'index

    # Afficher le nom du fichier CSV créé
    print(f"Données enregistrées dans {filename}")

# Appeler la fonction pour récupérer les données et les enregistrer dans un fichier CSV
print("Debut d'extration des")
crypto_actu()
print("extraction de donnée en 24h recus")
```
