```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger les données des deux fichiers CSV
df_2021 = pd.read_csv("resultats_election_2021.csv")
df_2024 = pd.read_csv("resultats_election_2024.csv")

# Afficher les données importées
print("Données de l'élection de 2021 :")
print(df_2021)
print("\nDonnées de l'élection de 2024 :")
print(df_2024)

# Exemple de visualisation (à adapter selon vos besoins)
plt.figure(figsize=(10, 6))

# Plot des résultats de l'élection de 2021
plt.bar(df_2021["Candidat"], df_2021["Voix"], color='blue', label='2021')

# Plot des résultats de l'élection de 2024
plt.bar(df_2024["Candidat"], df_2024["Voix"], color='red', label='2024')

plt.xlabel('Candidats')
plt.ylabel('Nombre de voix')
plt.title('Comparaison des résultats des élections de 2021 et 2024')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```

```python


# Calculer les totaux des pourcentages de voix pour chaque année
total_2021 = df_2021["%"].sum()
total_2024 = df_2024["%"].sum()

# Créer des labels pour les parts de voix
labels = ['2021', '2024']

# Créer une liste des totaux des pourcentages de voix
sizes = [total_2021, total_2024]

# Créer un diagramme à secteurs (pie chart)
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)

# Ajouter un titre
plt.title('Répartition des pourcentages de voix par année')

# Afficher le diagramme
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python


# Créer un subplot avec 2 lignes et 1 colonne
fig, axs = plt.subplots(2, 1, figsize=(10, 12))

# Plot des résultats de l'élection de 2021
axs[0].barh(df_2021["Candidat"], df_2021["%"], color='blue')
axs[0].set_xlabel('Pourcentage de voix (%)')
axs[0].set_title('Résultats des élections de 2021')
axs[0].invert_yaxis()  # Inverser l'ordre des candidats pour correspondre à l'ordre du DataFrame

# Plot des résultats de l'élection de 2024
axs[1].barh(df_2024["Candidat"], df_2024["%"], color='red')
axs[1].set_xlabel('Pourcentage de voix (%)')
axs[1].set_title('Résultats des élections de 2024')
axs[1].invert_yaxis()  # Inverser l'ordre des candidats pour correspondre à l'ordre du DataFrame

# Ajuster l'espacement entre les sous-graphiques
plt.tight_layout()

# Afficher le graphique
plt.show()
```

```python
def create_pie_chart(df, year):
    # Créer une figure et des axes
    fig, ax = plt.subplots()
    
    # Créer le diagramme de Kammerrer
    ax.pie(df["%"], labels=df["Candidat"], autopct='%1.1f%%', startangle=90)
    
    # Ajouter un titre
    ax.set_title(f"Répartition des pourcentages de voix - {year}")
    
    # Afficher le diagramme
    plt.show()

# Créer le diagramme de Kammerrer pour l'élection de 2021
create_pie_chart(df_2021, "2021")

# Créer le diagramme de Kammerrer pour l'élection de 2024
create_pie_chart(df_2024, "2024")
```

```python
# Concaténer les noms des candidats pour chaque élection
text_2021 = ' '.join(df_2021["Candidat"])
text_2024 = ' '.join(df_2024["Candidat"])

# Créer le nuage de mots pour l'élection de 2021
wordcloud_2021 = WordCloud(width=800, height=400, background_color='white').generate(text_2021)

# Créer le nuage de mots pour l'élection de 2024
wordcloud_2024 = WordCloud(width=800, height=400, background_color='white').generate(text_2024)

# Afficher les nuages de mots
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(wordcloud_2021, interpolation='bilinear')
plt.title('Nuage de mots - Élection de 2021')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(wordcloud_2024, interpolation='bilinear')
plt.title('Nuage de mots - Élection de 2024')
plt.axis('off')

plt.tight_layout()
plt.show()
```
